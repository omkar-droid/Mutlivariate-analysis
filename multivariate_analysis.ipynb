```python
from sklearn.datasets import load_iris
```

```python
import pandas as pd

# Load the dataset
iris = load_iris()

# Convert to a pandas DataFrame
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
# Add target to the DataFrame
iris_df['target'] = iris.target

# Display the DataFrame
print(iris_df.head())
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x=iris_df['petal length (cm)'], y= iris_df['petal width (cm)'], hue=iris_df['target'])
```

```python
import seaborn as sns

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Display the first few rows
print(titanic.head())
```

```python
sns.barplot(y=titanic['pclass'], x= titanic['sex'], hue=titanic['age'])
```

```python
sns.boxplot(x=titanic['sex'], y=titanic['age'], hue=titanic['survived'])
```

```python
sns.histplot(titanic[titanic['survived']==0]['age'], kde= True)
sns.histplot(titanic[titanic['survived']==1]['age'], kde= True)
```

```python
sns.heatmap(pd.crosstab(titanic['pclass'], titanic['survived']))
```

```python
sns.clustermap(pd.crosstab(titanic['parch'], titanic['survived']))
```
